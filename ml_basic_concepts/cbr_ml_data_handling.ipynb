```python
noisy_circles = datasets.make_circles(noise = 0.15)
```

```python
pyplot.figure(figsize=(5, 5))
pyplot.scatter(list(map(lambda x: x[0], noisy_circles[0])), list(map(lambda x: x[1], noisy_circles[0])), 
               c = noisy_circles[1], cmap = matplotlib.colors.ListedColormap(['red', 'grey']))
```

```python
def plot_2d_dataset(data, colors = ['red', 'grey']):
    pyplot.figure(figsize=(5, 5))
    pyplot.scatter(list(map(lambda x: x[0], data[0])), list(map(lambda x: x[1], data[0])), c = data[1], 
                   cmap = matplotlib.colors.ListedColormap(colors))
```

```python
plot_2d_dataset(noisy_circles)
```

## Matplotlib visualization <a name='visualization'></a>

```python
from pandas import DataFrame
```

```python
iris_frame = DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
```

```python
iris_frame.head()
```

```python
iris_frame.target = iris_frame.target.apply(lambda x : iris.target_names[x])
```

```python
iris_frame.head()
```

```python
iris_frame[iris_frame.target == 'setosa'].hist('sepal length (cm)')
```

```python
pyplot.figure(figsize=(20, 24))

plot_number = 0
for feature_name in iris['feature_names']:
    for target_name in iris['target_names']:
        plot_number += 1
        pyplot.subplot(4, 3, plot_number)
        pyplot.hist(iris_frame[iris_frame.target == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.xlabel('cm')
        pyplot.ylabel(feature_name[:-4])
```
